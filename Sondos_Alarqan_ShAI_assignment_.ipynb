**Task 2**

Calculate basic statistics mean, median,mode,minimum,and maximum salary, determine the range of salaries and find the standard deviation

```python
salary_column = df['TotalPay']

mean_salary = salary_column.mean()
median_salary = salary_column.median()
mode_salary = salary_column.mode().iloc[0]
min_salary = salary_column.min()
max_salary = salary_column.max()

salary_range = max_salary - min_salary

std_dev_salary = salary_column.std()

# the results
print(f"Mean Salary: {mean_salary}")
print(f"Median Salary: {median_salary}")
print(f"Mode Salary: {mode_salary}")
print(f"Minimum Salary: {min_salary}")
print(f"Maximum Salary: {max_salary}")
print(f"Salary Range: {salary_range}")
print(f"Standard Deviation of Salary: {std_dev_salary}")
```

Handle missing data by suitable method with explain why you use it

first i will drop Notes and status columns from the data frame
explanation : i use this way because both Column values are NaN so no need for them

```python
import pandas as pd

df = pd.read_csv("/content/Salaries.csv")
df.head()

df = df.drop(['Notes', 'Status'], axis=1)

#Display the modified DataFrame
print("DataFrame after dropping 'Notes' and 'Status' columns:")
print(df.head())

```

```python
# Save the modified DataFrame to a new CSV file
df.to_csv('/content/Salaries.csv', index=False)
```

secend step : Handling 'BasePay' , 'OvertimePay' , 'OtherPay'and  'benefits'
Since these columns represent numerical values related to pay, i consider imputing the missing values with the mean or median of the respective columns. This approach helps maintain the overall distribution of the pay data.

```python
import pandas as pd

df = pd.read_csv("/content/Salaries.csv")
df.head()

df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
print("DataFrame after filling 'BasePay,' 'OvertimePay,' and 'OtherPay'with the mean :")
print(df.head())
```

```python
# Save the modified DataFrame to a new CSV file
df.to_csv('/content/Salaries.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Visualization 1: Histogram for the distribution of salaries
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Pie chart for the proportion of employees in different departments
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(10)

plt.figure(figsize=(10, 10))
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```
